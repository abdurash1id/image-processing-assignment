```python
from PIL import Image
import matplotlib.pyplot as plt
img_pil = Image.open('images/test_image.jpg')
print(type(img_pil), img_pil.size, img_pil.mode)
plt.imshow(img_pil)
plt.axis('off')
```

```python
import numpy as np
img_np = np.array(img_pil)
print(img_np.shape, img_np.dtype)
```

```python
import matplotlib.image as mpimg
img_mpl = mpimg.imread('images/test_image.jpg')
print(img_mpl.shape)
plt.imshow(img_mpl)
plt.axis('off')
```

Matplotlib treats images as NumPy arrays because images are numerical pixel grids.

```python
from skimage import io, img_as_float
img_sk = io.imread('images/test_image.jpg')
img_float = img_as_float(img_sk)
print(img_float.min(), img_float.max())
plt.imshow(img_float)
plt.axis('off')
```

img_as_float rescales values safely to [0,1], unlike astype(float).

```python
import cv2
img_gray = cv2.imread('images/test_image.jpg', 0)
img_bgr = cv2.imread('images/test_image.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
```

```python
edges = cv2.Canny(img_gray, 100, 200)
plt.imshow(edges, cmap='gray')
plt.axis('off')
```

Reflection:
- Easiest: PIL
- Real-time: OpenCV
- RGB vs BGR: channel order
